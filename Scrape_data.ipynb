### Solo hemos extraido la data para una competencia, faltan varias 

> para prodecer hago el mismo procedimiento que hice para la copa america 2016, pero esta vez a todos los a;os que defini en la variable years

```python


def get_matches(year):
    web = f'https://en.wikipedia.org/wiki/{year}_Copa_Am%C3%A9rica'
    response = requests.get(web)
    content = response.text
    soup=BeautifulSoup(content, 'lxml')
    matches = soup.find_all('div', class_ = 'footballbox')
    home = []
    score = []
    away = []

    for match in matches: # necesito obtener de cada elemento el equipo visitante, el local y el resultado 
        home.append (match.find('th', class_='fhome').get_text())
        score.append(match.find('th', class_='fscore').get_text())
        away.append(match.find('th', class_='faway').get_text())
    
    dict_futbol = {'home': home,
              'score' : score,
              'away' : away
              }
    df_futbol = pd.DataFrame(dict_futbol)
    df_futbol['Year'] = year
    return df_futbol
```

```python
# En algunas tablas tengo problemas para extraer la informacion porque la pagina no tiene el mismo formato 
copa_df = [get_matches(year) for year in years]
```

```python
copa_df
```

##### Lo anterior me devolvio una lista con los resultados historicos como cada elemento. 

```python
# Concateno
df_america = pd.concat(copa_df, ignore_index= True)
```

```python
# Exporto a csv
df_america.to_csv('copa_america_resultados_historicos_data.csv', index= False)
```
