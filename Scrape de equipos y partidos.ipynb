# Extraer los demas equipos

```python
# Urls de los equipos 
team_urls
```

```python
# Nombre de los equipos 
equipos = ['Argentina','Canada','Chile','Peru',
           'Venezuela','Ecuador','Mexico','Jamaica',
           'Uruguay','Panama','United-States','Bolivia',
           'Colombia', 'Brazil', 'Costa-Rica', 'Paraguay']
```

```python
# DataFrame vacío donde se almacenará la información combinada
df_completo = pd.DataFrame()
```

```python
# Iterar sobre cada URL y nombre de animal
for url, equipo in zip(team_urls, equipos):
    try:
        # Leer la primera tabla de la URL especificada
        headers = {"User-Agent": "pandas"}
        df = pd.read_html(url, storage_options=headers,attrs={'id': 'stats_standard_685'})[0]
        
        
        # Agregar una columna con el nombre del equipo
        df['pais'] = equipo
       
                      
        # Concatenar al DataFrame completo
        df_completo = pd.concat([df_completo, df], ignore_index=True)
        
            
         # Hacer una pausa de 5 segundos entre cada solicitud
        time.sleep(5)
        
    except Exception as e:
        print(f"No se pudo procesar la URL {url}: {str(e)}")
```

```python
df_completo
```

```python
# Como tiene multiindice, dropeo el primer nivel
df_completo.columns = df_completo.columns.droplevel(level=0)
```

```python
df_completo = df_completo.rename(columns={df_completo.columns[33]: 'pais'})
```

```python
# Dropeo lor registros que tienen los totales, que tiene uno por equipo 

df_completo = df_completo[~((df_completo['Player'] == 'Squad Total') | (df_completo['Player'] == 'Opponent Total'))]
```

```python
# COntrolo que cada equipo tenga numero parecidos de jugadores
df_completo['pais'].value_counts()
```

```python
df_completo
```

```python
# Me quedo con las columnas que me interesan
df_limpio = df_completo[cols]
#Limpio los nombres de las columnas para eliminar, mayusculas o espacios
df_limpio = clean_names(df_limpio)
# Limpio el nombre de los jugadores

df_limpio['player'] = df_limpio['player'].str.strip()  # Eliminar espacios en blanco al principio y al final
df_limpio['player'] = df_limpio['player'].str.lower()  # Convertir a minúsculas
df_limpio['player'] = df_limpio['player'].str.replace('[^\w\s]', '')  # Eliminar caracteres especiales
df_limpio['player'] = df_limpio['player'].apply(lambda x: unidecode(x))
```

### Scraping data de los demas equipos temporada por temporada

```python
# Que temporada es
season = [2024]
```

```python
all_matches = []
```

```python
team_urls
```

```python
import time

for url, year in zip(team_urls, season):
    
    
    for team_url in team_urls:
        team_name = team_url.split('/')[-1].replace('-Stats','').replace('-Men',' ')
        
        data = requests.get(team_url)
        matches = pd.read_html(data.text, match="Scores & Fixtures")[0]
        
        soup = BeautifulSoup(data.text)
        links = [l.get('href') for l in soup.find_all('a')]
        links = [l for l in links if l and 'all_comps/shooting/' in l]
        data = requests.get(f'https://fbref.com{links[0]}')
        shooting = pd.read_html(data.text, match="Shooting")[0]
        shooting.columns = shooting.columns.droplevel()
        
        try:
            team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", 
                                                "PK", "PKatt"]], on = "Date")
        except ValueError:
            continue # salta si estos datos no estan disponibles
            
        
        team_data['Season'] = year
        team_data['Team'] = team_name
        all_matches.append(team_data)
        time.sleep(3) # no hace nada durante tres segundos para que las paginas no nos corran al hacer el scrapping
        

    
```
